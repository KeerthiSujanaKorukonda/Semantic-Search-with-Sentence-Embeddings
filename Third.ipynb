```python
import json

sentence_embeddings = []

# Compute embeddings for each sentence and store them in a list
for sentence in sentences[:100]:
    sentence_embedding = model.encode(sentence)  # Assuming model.encode returns a list of embeddings
    sentence_embeddings.append(sentence_embedding)
    # break

# Open the JSON file in write mode
with open("embeddings.json", "w") as f:
    # Convert the embeddings list to a list of lists
    embeddings_list = [embedding.tolist() for embedding in sentence_embeddings]

    # Write the embeddings to the JSON file
    json.dump(embeddings_list, f)



print(len(sentence_embeddings))

```

```python
def query(sent):
  embedding = model.encode(sent)
  embedding = embedding.tolist()
  results = collection.query(
    query_embeddings=embedding,  # No specific query embeddings provided
    n_results=3,
    where={},              # No metadata filter applied
    where_document={},     # No document content filter applied
    include=["embeddings","documents"]        # Retrieve up to 10 results
   # No document content filter applied
)
  return results
```

```python
print(query('Global warming is driving polar bears toward extinction'))
```
